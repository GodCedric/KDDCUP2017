### 录入数据

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from datetime import datetime,timedelta,date,time
```

```python
travel_time_data = pd.read_csv('travel_time_raw_data.csv')
volume_data = pd.read_csv('volume_raw_data.csv')
```

### 平均时间数据缺失值，异常值处理 

```python
travel_time_data
```

```python
# 风向异常值处理，以最近的风向值代替
wind_direction = travel_time_data['wind_direction']
```

```python
wind_direction[wind_direction > 360] = np.nan
```

```python
wind_direction.fillna(method='ffill', inplace=True)
```

```python
# last_20min值缺失暂时还是以最近的补充吧
last_20min = travel_time_data['last_20min']
last_20min.fillna(method='ffill', inplace=True)
# 最开始的缺失值用后面临近的处理
last_20min.fillna(method='bfill', inplace=True)
```

```python
travel_time_data
```

```python
# 写出数据
travel_time_data.to_csv('travel_time_data.csv', index=False)
```
