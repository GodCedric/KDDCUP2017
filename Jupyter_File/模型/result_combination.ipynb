```python
time_result_1 = mean_time_result['avg_travel_time']
time_result_2 = xgboost_time_result['avg_travel_time']
#time_result_3 = kNN_time_result['avg_travel_time']
time_result_4 = RF_time_result['avg_travel_time']
#time_result_5 = xgboost_time_result2['avg_travel_time']

time_result = 0.25*time_result_1 + 0*time_result_2 + 0.25*time_result_4 

#volume_result_1 = mean_volume_result['volume']
#volume_result_2 = xgboost_volume_result['volume']
#volume_result_3 = kNN_volume_result['volume']
#volume_result_4 = RF_volume_result['volume']
#volume_result_5 = xgboost_volume_result2['volume']

volume_result = 0.25*volume_result_1 + 0*volume_result_2 + 0.25*volume_result_3 + 0.25*volume_result_4 + 0.25*time_result_5
```

```python
time_result_1 = mean_time_result['avg_travel_time']
time_result_2 = xgboost_time_result['avg_travel_time']
time_result_3 = kNN_time_result['avg_travel_time']
time_result_4 = RF_time_result['avg_travel_time']
#time_result_5 = xgboost_time_result2['avg_travel_time']

time_result = (1*time_result_1 + 1*time_result_2 + 1*time_result_3 + 1*time_result_4 )/4

volume_result_1 = mean_volume_result['volume']
volume_result_2 = xgboost_volume_result['volume']
volume_result_3 = kNN_volume_result['volume']
volume_result_4 = RF_volume_result['volume']
#volume_result_5 = xgboost_volume_result2['volume']

volume_result = (1*volume_result_1 + 1*volume_result_2 + 1*volume_result_3 + 1*volume_result_4)/4
```

```python
time_result_1 = mean_time_result['avg_travel_time']
time_result_2 = xgboost_time_result['avg_travel_time']
#time_result_3 = kNN_time_result['avg_travel_time']
time_result_4 = RF_time_result['avg_travel_time']

# 精细化融合
A2_result = 0*time_result_1[0:84] + 1*time_result_2[0:84] + 0*time_result_4[0:84]
A3_result = 0*time_result_1[84:168] + 1*time_result_2[84:168] + 0*time_result_4[84:168]
B1_result = 0*time_result_1[168:252] + 1*time_result_2[168:252] + 0*time_result_4[168:252]
B3_result = 0.5*time_result_1[252:336] + 0.25*time_result_2[252:336] + 0.25*time_result_4[252:336]
C1_result = time_result_2[336:420]
C3_result = 0.6*time_result_2[420:504] + 0.4*time_result_4[420:504]
```

```python
time_predict = np.concatenate([A2_result,A3_result,B1_result,B3_result,C1_result,C3_result], axis=0)
```

```python
travel_time_submission['avg_travel_time'] = time_predict
```
